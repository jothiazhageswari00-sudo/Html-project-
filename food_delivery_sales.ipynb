```python
import pandas as pd
import matplotlib.pyplot as plt

# Create Data
data = {
    "Order_Time": ["Breakfast", "Lunch", "Dinner", "Lunch", "Evening",
                   "Dinner", "Lunch", "Breakfast", "Dinner", "Lunch"],
    "Location": ["Madurai", "Chennai", "Madurai", "Trichy", "Chennai",
                 "Madurai", "Chennai", "Trichy", "Madurai", "Chennai"],
    "Food_Item": ["Dosa", "Biryani", "Pizza", "Biryani", "Burger",
                  "Pizza", "Biryani", "Idli", "Pizza", "Burger"]
}

df = pd.DataFrame(data)

# Display Data
print("Food Delivery Data")
print(df)

# Peak Order Times
print("\nPeak Order Times:")
print(df["Order_Time"].value_counts())

# Most Popular Dishes
print("\nMost Popular Dishes:")
print(df["Food_Item"].value_counts())

# Orders by Location
print("\nOrders by Location:")
print(df["Location"].value_counts())

# Dashboard

# Popular Food Items
plt.figure(figsize=(6,4))
df["Food_Item"].value_counts().plot(kind="bar")
plt.title("Popular Food Items")
plt.xlabel("Food Item")
plt.ylabel("Orders")
plt.show()

# Peak Order Times
plt.figure(figsize=(6,4))
df["Order_Time"].value_counts().plot(kind="bar")
plt.title("Peak Order Times")
plt.xlabel("Order Time")
plt.ylabel("Orders")
plt.show()

# Orders by Location
plt.figure(figsize=(5,5))
df["Location"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Orders by Location")
plt.ylabel("")
plt.show()
```
