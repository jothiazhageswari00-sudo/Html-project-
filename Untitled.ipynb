```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
"Student_ID": [101, 102, 103, 104, 105, 106, 107, 108],
"Student_Name": ["Arun", "Bala", "Chitra", "Divya", "Eshan", "Farah", "Gokul", "Hema"],
"Subject": ["Maths", "Science", "English", "Maths", "Science", "English", "Maths", "Science"],
"Marks": [78, 35, 62, 90, 44, 28, 55, 39],
"Attendance": [92, 75, 88, 95, 80, 70, 85, 78]
}
df = pd.DataFrame(data)
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
print(df.head())
print("Average Marks:", df["Marks"].mean())
print("Highest Marks:", df["Marks"].max())
print("Lowest Marks:", df["Marks"].min())
print(df["Result"].value_counts())
plt.bar(df["Student_Name"], df["Marks"])
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.title("Student vs Marks")
plt.show()
plt.pie(df["Result"].value_counts(), labels=df["Result"].value_counts().index, autopct='%1.1f%%')
plt.title("Pass vs Fail")
plt.show()
```
